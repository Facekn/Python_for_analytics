```python
import pandas as pd
import matplotlib.pyplot as plt

# Построение круговой диаграммы для распределения домов от наличия вида на набережную
df_house_data = pd.read_csv("kc-house-data (1) (2).csv")
waterfront_counts = df_house_data['waterfront'].value_counts()
plt.figure(figsize=(2, 2))
plt.pie(waterfront_counts, labels=['Нет', 'Да'], autopct='%1.1f%%', startangle=90)
plt.title('Распределение домов от наличия вида на набережную')
plt.axis('equal')  # Установка равного масштаба по осям x и y для круговой диаграммы
plt.show()
```

```python
# Построение гистограммы для распределения этажей домов
df_house_data['floors'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.title('Распределение этажей домов')
plt.show()
```

```python
import seaborn as sns
# Установка стиля Seaborn
sns.set(style="whitegrid")
```

```python
# Изучите распределение домов от наличия вида на набережную (Pie chart)
plt.figure(figsize=(1, 1))
df_house_data['waterfront'].value_counts().plot(kind='pie', labels=['Нет', 'Да'], autopct='%1.1f%%', startangle=90)
plt.title('Распределение домов от наличия вида на набережную')
plt.axis('equal')
plt.show()
```

```python
# Изучите распределение этажей домов (Histogram)
plt.figure(figsize=(2, 3))
sns.histplot(data=df_house_data, x='floors', bins=20, kde=False)
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.title('Распределение этажей домов')
plt.show()
```

```python
plt.figure(figsize=(2, 3))
sns.countplot(data=df_house_data, x='condition')
plt.xlabel('Состояние дома')
plt.ylabel('Количество домов')
plt.title('Распределение состояния домов')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_house_data, x='condition', hue='floors')
plt.xlabel('Состояние дома')
plt.ylabel('Количество домов')
plt.title('Многорядная столбчатая диаграмма для состояния и этажей')
plt.legend(title='Количество этажей')
plt.show()
```

## Выводы:
Круговая диаграмма (Pie chart) показывает, что большинство домов не имеют вид на набережную, в то время как домов с видом на набережную существенно меньше.

Гистограмма (Histogram) демонстрирует распределение домов по количеству этажей, позволяя увидеть, Что наибольшее количество домов - одноэтажные.

График столбчатой диаграммы (Bar plot) демонстрирует распределение домов по разным уровням состояния, показывая, что большинство домов находится в удовлетворительном состоянии.

```python
# 1. График зависимости стоимости от квадратуры жилой площади (scatter plot)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_house_data, x='sqft_living', y='price', alpha=0.7)
plt.xlabel('Квадратура жилой площади (sqft)')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Зависимость стоимости от квадратуры жилой площади')
plt.show()
```

##Выводы:

Из графика видно, что есть положительная корреляция между квадратурой жилой площади и стоимостью недвижимости - чем больше площадь дома, тем выше его стоимость. Что в принципе вполне закономерно

```python
# 2. График распределения стоимости недвижимости по количеству комнат (box plot)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_house_data, x='bedrooms', y='price')
plt.xlabel('Количество комнат')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Распределение стоимости недвижимости по количеству комнат')
plt.show()
```

##Выводы:

График показывает, что стоимость недвижимости варьируется в зависимости от количества комнат в доме. Дома с большим количеством комнат могут иметь более высокую стоимость, но есть также выбросы с низкой стоимостью для некоторых комбинаций количества комнат.

```python
# 3. График распределения стоимости недвижимости по виду на набережную и количеству этажей (bar plot)
plt.figure(figsize=(10, 6))
sns.barplot(data=df_house_data, x='waterfront', y='price', hue='floors')
plt.xlabel('Вид на набережную')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Распределение стоимости недвижимости по виду на набережную и этажам')
plt.legend(title='Количество этажей')
plt.show()
```

##Выводы:

Этот график позволяет сравнить среднюю стоимость недвижимости с видом на набережную и без него, а также учитывает влияние количества этажей. На графике видно, что дома с видом на набережную и несколькими этажами имеют более высокую среднюю стоимость.

```python
# 4. График распределения стоимости недвижимости в зависимости от года постройки (box plot)
plt.figure(figsize=(20, 8))
sns.boxplot(data=df_house_data, x='yr_built', y='price')
plt.xlabel('Год постройки')
plt.ylabel('Стоимость недвижимости (миллион долларов)')
plt.title('Распределение стоимости недвижимости по году постройки')
plt.xticks(rotation=90)
plt.show()
```

##Выводы:

График позволяет увидеть распределение стоимости недвижимости по годам постройки. Старые дома имеют разнообразные цены, но есть некоторые выбросы с высокой стоимостью. Дома, построенные недавно, имеют более высокую среднюю стоимость.

```python
# Построение тепловой карты
numeric_columns = df_house_data.select_dtypes(include=['int64', 'float64'])
corr_matrix = numeric_columns.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Тепловая карта влияния характеристик на стоимость жилья')
plt.show()
```
